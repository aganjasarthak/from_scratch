```python

# Import required libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Step 1: Simulate a dataset
# Generate a sample dataset
data = {
    'x1': [1, 2, 3, 4, 5],
    'x2': [5, 4, 3, 2, 1],
    'x3': [2, 3, 4, 5, 6],
    'y': [0, 1, 0, 1, 0]
}
df = pd.DataFrame(data)

# Step 2: Define feature subsets
feature_subsets = [['x1', 'x2'], ['x1', 'x3'], ['x2', 'x3']]

# Initialize an empty list to store predictions for aggregation
all_predictions = []

# Step 3: Train decision trees for each subset and make predictions
for subset in feature_subsets:
    # Extract features and target
    X = df[subset]
    y = df['y']

    # Train Decision Tree with max_depth=10
    clf = DecisionTreeClassifier(random_state=42, max_depth=10)
    clf.fit(X, y)

    # Predict using the same dataset
    predictions = clf.predict(X)
    all_predictions.append(predictions)  # Store predictions for aggregation

    # Plot the decision tree
    plt.figure(figsize=(10, 6))
    plot_tree(clf, feature_names=subset, class_names=['0', '1'], filled=True)
    plt.title(f"Decision Tree for subset: {subset}")
    plt.show()

# Step 4: Aggregate predictions by averaging
# Convert predictions list to a DataFrame for easier computation
prediction_df = pd.DataFrame(all_predictions).T  # Transpose to align rows
prediction_df.columns = [f"Subset_{i+1}" for i in range(len(feature_subsets))]

# Compute the average prediction for each row
prediction_df['Aggregated_Prediction'] = prediction_df.mean(axis=1)

# Add true labels for comparison
prediction_df['True_Label'] = df['y']

# Display results
print("\nAggregated Predictions:")
print(prediction_df)
```
